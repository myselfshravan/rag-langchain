```python
from langchain.schema import Document
from langchain.text_splitter import RecursiveCharacterTextSplitter


def split_text(documents: list[Document]):
    custom_text_splitter = RecursiveCharacterTextSplitter(chunk_size=200, chunk_overlap=200, separators=["\n"],
                                                          length_function=len)
    chunks = custom_text_splitter.split_documents(documents)
    print(f"Number of documents: {len(documents)}  into {len(chunks)} chunks")

    document = chunks[2]
    print(document.page_content)
    print(document.metadata)

    return chunks


new_chunks = split_text(docs)
```

```python
from langchain_community.vectorstores import Chroma
from langchain_openai import OpenAIEmbeddings
from config import openai_api_key

CHROMA_PATH = "chroma"


def save_to_chroma(chunks: list[Document]):
    db = Chroma.from_documents(
        chunks, OpenAIEmbeddings(openai_api_key=openai_api_key), persist_directory=CHROMA_PATH
    )
    db.persist()
    print(f"Saved {len(chunks)} chunks to {CHROMA_PATH}.")
```

```python
embedding_function = OpenAIEmbeddings(openai_api_key=openai_api_key)
db = Chroma(persist_directory=CHROMA_PATH, embedding_function=embedding_function)

results = db.similarity_search_with_relevance_scores("From 1972 – 1982", k=3)
```
